## Data Visualization With Matplotlib

Matplotlib is a powerful plotting library for python that enables the creation of static, animated, and interactive visualizations. It is widely used for data visualization in data science and analytics. In this lesson, we will cover the basics of Matplotlib, including creating various type of plots and customizing them.

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

## create a line plot 
plt.plot(x,y,)
plt.xlabel('x-axix')
plt.ylabel('y Axis')
plt.title('basic of Line Plot')
```

```python
## Create a customized line plot

plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=1,markersize=9)
plt.grid(True)
```

```python
## Multiple plots
## sample data 
x=[x for x in range(1,6)]
y1=[x*x for x in range(1,6)]
y2=[x for x in range(1,6)]


plt.figure(figsize=(9,5))

plt.subplot(1,2,1)
plt.plot(x,y1,color='green')
plt.title('Plot1')

plt.subplot(1,2,2)
plt.plot(x,y2,color='red')
plt.title('Plot2')

```

```python
## bar plot
categories=['a','b','c','d','e']
values=[5,7,8,3,6]

## create a bar plot 
plt.bar(categories,values,color='purple')
```

## Histograms 

Histograms are used to represent the distribution of a dataset. they divide the data into bins and count the number of data points in each bin.

```python
data=[1,2,2,2,3,3,3,4,4,4,4,5,5]

# create a histogram 
plt.hist(data,bins=5,color='orange',edgecolor='black')

```

```python
# create a scatter plot 
plt.scatter(x,y,color='blue',marker='x')
```

```python
## pipe chart 
labls=['a','b','c','d']
sizes=[30,20,40,10]
color=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.2,0.1,0.1,0.1)


## create 
plt.pie(sizes,explode=explode,labels=labls,colors=color,shadow=True)
```
