## Linear Regression, Ridge And Lasso 

```python
## House Pricing dataset 

import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('train.csv')
```

```python
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data.head()
```

```python
## dividing the dataset into indemendent and dependent features
test_data=pd.read_csv('test.csv')
x=data.iloc[:,:-1]  ## independent features 

y=data.iloc[:,-1] ## Dependent features
```

```python
test_data.head()
x_test=test_data.iloc[:,:-1]
y_test=test_data.iloc[:,-1]
```

```python
x_test.head()
```

```python
y_test.head()
```

```python
x.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
lin_reg=LinearRegression()
```

```python
mse=cross_val_score(lin_reg,x,y,scoring='neg_mean_squared_error',cv=5)
```

```python
print(mse)
```

```python
mean_mse=np.mean(mse)
print(mean_mse)
```

```python
# lin_reg.predict()
```

```python
### Ridge Regression 

from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge=Ridge()

params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,20]}

ridge_regressor=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)

ridge_regressor.fit(x,y)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
## Lasso Regression 
from sklearn.linear_model import Lasso
lasso=Lasso()
```

```python
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}

lasso_regressor=GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(x,y)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)

```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
## Linear Regression

lin_reg=LinearRegression()
lin_reg.fit(x_train,y_train)
mse=cross_val_score(lin_reg,x_train,y_train,scoring='neg_mean_squared_error',cv=5)

mean_mse=np.mean(mse)
print(mean_mse)
```

```python
## Ridge Regression with x_train and y_train 

ridge_regressor.fit(x_train,y_train)
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
## Lasso Regression with x_train and y_train 
lasso_regressor.fit(x_train,y_train)
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
# Score of the Lasso regression 

y_pred=lasso_regressor.predict(x_test)

from sklearn.metrics import r2_score

r2_score1=r2_score(y_pred,y_test)
print(r2_score1)
```

```python
# Score of the Ridge Regression 

y_pred=ridge_regressor.predict(x_test)

r2_score1=r2_score(y_pred,y_test)
print(r2_score1)
```

```python
# Score of the Linear Regression 

y_pred=lin_reg.predict(x_test)

r2_score1=r2_score(y_pred,y_test)
print(r2_score1)
```
